```python
X,Y=make_classification(n_samples=1000,n_features=25,n_informative=3,n_redundant=22,weights=[0.9, 0.1],random_state=1441)
```

```python
X_noisy = X + np.random.normal(0, 1.0, size=X.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_noisy,Y,test_size=0.3,random_state=2)
```

```python
train=[]
test=[]
iter=10
for i in range(1, iter+1):
  model1=tree.DecisionTreeClassifier(max_depth=i)
  model1.fit(x_train,y_train)

  y_traintest=model1.predict(x_train)
  train_acc=accuracy_score(y_train,y_traintest)
  train.append(train_acc)

  y_pred=model1.predict(x_test)
  test_acc=accuracy_score(y_test,y_pred)
  test.append(test_acc)

  print("Train Accuracy Iter ",i,": ",train_acc,end="\t\t\t")
  print("Test Accuracy Iter ",i,": ",test_acc)
```

```python
plt.plot(train, '-o', label='Train')
plt.plot(test, '-o', label='Test')
plt.ylim(0.75, 1)
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy')
plt.legend()
plt.show()
```
